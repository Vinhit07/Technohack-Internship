## Checking & Removing Outliers in Age Column

```python
Q1 = df.Age.quantile(0.25)
Q3 = df.Age.quantile(0.75)
IQR = Q3 - Q1
ll = Q1 - 1.5 * IQR
ul = Q3 + 1.5 * IQR
print(ul,ll)
```

```python
df = df[(df['Age'] >= ll) & (df['Age'] <= ul)]
#df = df[(df['Age']<60)]
df
```

## Converting Null Values in Age Column 

```python
df['Age'].mean()
```

## After Treating Outliers

```python
sns.boxplot(x=df['Age'],showfliers=True) 
```

```python
df['Age'].fillna(df['Age'].mean())
```

## Checking & Removing Outliers in Fare Column

```python
sns.boxplot(x=df['Fare'],showfliers=True) 
```

```python
Q1 = df.Fare.quantile(0.25)
Q3 = df.Fare.quantile(0.75)
IQR = Q3 - Q1
ll = Q1 - 1.5 * IQR
ul = Q3 + 1.5 * IQR
print(ul,ll)
```

```python
mean = df['Fare'].mean()
df.loc[(df['Fare'] > ul) | (df['Fare'] < ll), 'Fare'] = mean
```

```python
sns.boxplot(x=df['Fare'],showfliers=True) 
```
